```python
X=df[['Age','DistanceFromHome','JobSatisfaction','WorkLifeBalance','EnvironmentSatisfaction','PerformanceRating']].values
X[0:5]
```

```python
y=df[['Attrition']].values
y[0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

```python
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train))*100)
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat)*100)
```

```python
Ks = 5
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks): 
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
m_list=mean_acc.tolist()
mean=0
for i in m_list:
    i=i*100
    mean=mean+i
mean=mean/4
print("Mean Accuracy of test data:",mean)
```
